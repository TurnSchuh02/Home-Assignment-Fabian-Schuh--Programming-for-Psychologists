### Loading data sets
Next, the code should be able to locate relevant files for brain activation visualization (anatomical & functional) and histogram generation (functional). These files are identifyable through their file type (.nii.gz), and distinguishable through components of their title (anatomical: anatomical; functional: coactivation). Finally, loading the located files is necessary.

```python
# Defining directory where files are located
data_dir = "/Users/lukas/Desktop/Python/Home-Assignment-Fabian-Schuh--Programming-for-Psychologists"

# Locating relevant files & providing feedback
files = [f for f in os.listdir(data_dir) if f.endswith(".nii.gz")]
print("Found files:", files)

# Assigning files based on naming  (anatomical & coactivation) & providing feedback
anatomical_file = next(f for f in files if "anatomical" in f.lower())
functional_file = next(f for f in files if "coactivation" in f.lower())
print("Anatomical file:", anatomical_file)
print("Functional file:", functional_file)

# Loading located files
anatomical_img = nib.load(os.path.join(data_dir, anatomical_file))
functional_img = nib.load(os.path.join(data_dir, functional_file))
```

### Brain Map Visualisation
The below function visualisez the anatomical brain data and overlays anatomical data to show activation. Cumstomization supports understandability of the visualized data (title, colorbar, colormap). Additionally, exclusion of low values allows for focussing on relevantly high activation while specifying coordinates allows for visualization of targeted sections of the brain. 

```python
# Visualizing functional data of anatomical scan
plotting.plot_stat_map(
    functional_img,
    bg_img=anatomical_img,
    colorbar = True,                                                                    # Colorbar display
    title=f"Functional Activation Map: Brain Activity - {concept}",                     # Title including respective concept
    threshold=0.1,                                                                      # Excluding low values
    cmap="inferno",                                                                     # Colormap
    cut_coords=(6, -2, 0)                                                               # Brain coordinates
)
plotting.show()
```

### Histogram: Distribution of Values in Functional Data File
Visualization of value distribution in the uploaded functional data file allows for insights into brain activation. Excluding negative values makes sure only activation is depicted.

```python
# Loading functional data as array
functional_data = functional_img.get_fdata()

# Extracting positive values
positive_values = functional_data[functional_data > 0]

# Plotting histogram
plt.hist(positive_values, bins=50, color="orange", edgecolor="yellow")
plt.title("Histogram of Functional Data Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
